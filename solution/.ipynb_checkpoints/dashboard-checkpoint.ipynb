# Import Data

```python
# Import the CSVs to Pandas DataFrames
file_path = Path("Data/sfo_neighborhoods_census_data.csv")
sfo_data = pd.read_csv(file_path, index_col="year")

file_path = Path("Data/neighborhoods_coordinates.csv")
df_neighborhood_locations = pd.read_csv(file_path)
```

- - -

## Panel Visualizations

In this section, you will copy the code for each plot type from your analysis notebook and place it into separate functions that Panel can use to create panes for the dashboard. 

These functions will convert the plot object to a Panel pane.

Be sure to include any DataFrame transformation/manipulation code required along with the plotting code.

Return a Panel pane object from each function that can be used to build the dashboard.

Note: Remove any `.show()` lines from the code. We want to return the plots instead of showing them. The Panel dashboard will then display the plots.

```python
# Define Panel Visualization Functions
def housing_units_per_year():
    """Housing Units Per Year."""
    
    sfo_mean_units_year = (
    sfo_data[["housing_units"]].groupby("year").mean().sort_values(["year"])
    )
    
    housing_units_plot = sfo_mean_units_year.hvplot(
    kind="bar",
    title="Housing Units in San Francisco from 2010-2014",
    ylim=(372000,387500), 
    ylabel="Housing Units",
    xlabel="Year",
    figsize=(20,10),
    rot=45
    );
    
    return housing_units_plot


def average_gross_rent():
    """Average Gross Rent in San Francisco Per Year."""
    
    avg_gross_rent = sfo_data[["sale_price_sqr_foot", "gross_rent"]].groupby("year").mean().sort_values("year")

    avg_gross_rent_plot = avg_gross_rent.hvplot.line(
    title="Average Gross Rent per Year in San Francisco",  
    y="gross_rent", 
    ylabel="Gross Rent",
    xlabel="Year",
    figsize=(20,10),
    rot=45
    );
    
    return avg_gross_rent_plot


def average_sales_price():
    """Average Sales Price Per Year."""
    
    avg_gross_rent = sfo_data[["sale_price_sqr_foot", "gross_rent"]].groupby("year").mean().sort_values("year")
    
    avg_sales_price_plot = avg_gross_rent.hvplot.line(
    title="Average Sales Price per Sq. Ft. per Year in San Francisco",  
    y="sale_price_sqr_foot",
    ylabel="Avg. Sale Price per Sq. Ft.",
    xlabel="Year",
    figsize=(20,10),
    rot=45
    );

    return avg_sales_price_plot


def average_price_by_neighborhood():
    """Average Prices by Neighborhood."""
    
    neighborhood_df = sfo_data.groupby(["year", "neighborhood"]).mean()
    
    avg_price_by_hood_plot = neighborhood_df["sale_price_sqr_foot"].hvplot.line(
    title="Avg. Price per Sq. Ft. of Neighborhoods in San Francisco",
    x="year",
    xlabel="Year",
    ylabel="Sale Price per Sq. Ft.",
    label=" ",
    groupby="neighborhood",
    rot=45,
    ).opts(yformatter="%.0f")
    
    return avg_price_by_hood_plot


def top_most_expensive_neighborhoods():
    """Top 10 Most Expensive Neighborhoods."""
    
    hood_prices = sfo_data[
    ["neighborhood", "sale_price_sqr_foot", "housing_units", "gross_rent"]
    ].groupby("neighborhood").mean()

    top_10_hood_prices = hood_prices.sort_values("sale_price_sqr_foot", ascending=False).head(10).reset_index()
    
    top_10_hood_prices_plot = top_10_hood_prices.hvplot.bar(
    title="Top 10 Most Expensive Neighborhoods in SF",
    y="sale_price_sqr_foot",
    ylabel="Avg. Sale Price per Sq. Ft.",
    x="neighborhood",
    xlabel="Neighborhood",
    ylim=(600,950),
    rot=45
    )
    
    return top_10_hood_prices_plot

def parallel_coordinates():
    """Parallel Coordinates Plot."""
    
    hood_prices = sfo_data[
    ["neighborhood", "sale_price_sqr_foot", "housing_units", "gross_rent"]
    ].groupby("neighborhood").mean()
    
    top_10_hood_prices = hood_prices.sort_values("sale_price_sqr_foot", ascending=False).head(10).reset_index()
    
    par_coor_plot = px.parallel_coordinates(
    top_10_hood_prices,
    dimensions=["sale_price_sqr_foot", "housing_units", "gross_rent"],
    color="sale_price_sqr_foot",
    )
    
    return par_coor_plot


def parallel_categories():
    """Parallel Categories Plot."""
    
    hood_prices = sfo_data[
    ["neighborhood", "sale_price_sqr_foot", "housing_units", "gross_rent"]
    ].groupby("neighborhood").mean()
        
    top_10_hood_prices = hood_prices.sort_values("sale_price_sqr_foot", ascending=False).head(10).reset_index()
    
    par_cat_plot = px.parallel_categories(
    top_10_hood_prices,
    dimensions=["neighborhood", "sale_price_sqr_foot", "housing_units", "gross_rent"],
    color="sale_price_sqr_foot",
    )
    
    return par_cat_plot


def neighborhood_map():
    """Neighborhood Map"""
    
    hood_prices = sfo_data[
    ["neighborhood", "sale_price_sqr_foot", "housing_units", "gross_rent"]
    ].groupby("neighborhood").mean()
    mean_hood_values = hood_prices.sort_values("neighborhood", ascending=True).reset_index()
    hood_locale_value = pd.concat((mean_hood_values, df_neighborhood_locations), axis=1, join="inner")
    
    hood_map_scatter = px.scatter_mapbox(
    hood_locale_value,
    title="Average Sale Price per Sq. Ft. and Gross Rent",
    lat="Lat",
    lon="Lon",
    size="sale_price_sqr_foot",
    color="gross_rent",
    zoom=10.5
    )
    
    return hood_map_scatter

```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
welcome_column = pn.Column(
    "## Real Estate Analysis of San Francisco from 2010-2016",
    "#### Welcome to the San Francisco neighborhood analysis dashboard for 2010-2016. Use the tabs above to navigate through the different sections.",
    neighborhood_map()
)

market_analysis = pn.Row(
    housing_units_per_year(),
    average_gross_rent(),
    average_sales_price()
)

hood_analysis = pn.Column(
    average_price_by_neighborhood(),
    top_most_expensive_neighborhoods()
)

parallel_plots_analysis = pn.Column(
    parallel_coordinates(),
    parallel_categories()
)
```

```python
rental_analysis_dashboard = pn.Tabs(
    (
        "Welcome",
        welcome_column
    ),
    (
        "Yearly Market Analysis",
        market_analysis
    ),
    (
        "Neighborhood Analysis",
        hood_analysis
    ),
    (
        "Parallel Plots Analysis",
        parallel_plots_analysis
    )
)

rental_analysis_dashboard
```

## Serve the Panel Dashboard

```python
rental_analysis_dashboard.servable()
```
